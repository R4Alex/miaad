```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci칩n y su derivada (gradiente)
def f(x):
    return x**2 + 4*x + 4

def grad_f(x):
    return 2*x + 4

# Par치metros del gradiente descendente
x_vals = [0]  # punto inicial
learning_rate = 0.1
n_iter = 10

# Ejecutar gradiente descendente
x = x_vals[0]
for _ in range(n_iter):
    grad = grad_f(x)
    x = x - learning_rate * grad
    x_vals.append(x)

# Preparar la gr치fica
x_plot = np.linspace(-6, 2, 400)
y_plot = f(x_plot)

# Graficar la funci칩n
plt.figure(figsize=(8, 6))
plt.plot(x_plot, y_plot, label='f(x) = x + 4x + 4')
plt.title("Gradiente descendente paso a paso")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)

# Dibujar los pasos
for i in range(len(x_vals)-1):
    x0 = x_vals[i]
    x1 = x_vals[i+1]
    plt.plot(x0, f(x0), 'ro')
    plt.annotate('', xy=(x1, f(x1)), xytext=(x0, f(x0)),
                 arrowprops=dict(arrowstyle='->', color='green', lw=2))

# Punto final
plt.plot(x_vals[-1], f(x_vals[-1]), 'bo', label='M칤nimo aproximado')
plt.legend()
plt.tight_layout()
plt.show()
```
